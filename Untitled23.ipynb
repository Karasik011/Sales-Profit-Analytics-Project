```python
from textwrap import fill
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount("/content/drive")

%cd /content/drive/MyDrive/Projects

countries = pd.read_csv("countries.csv")
events = pd.read_csv("events.csv")
products = pd.read_csv("products.csv")


for df in [countries, events, products]:
    df.replace(['', ' ', 'nan', 'None', 'NaN'], np.nan, inplace=True)

print(countries.isna().sum()[countries.isna().sum() > 0])
print(events.isna().sum()[events.isna().sum() > 0])
print(products.isna().sum()[products.isna().sum() > 0])


#Fill nan - in countries we fill with region or code(find it in Google), for events we fill units sold with previous value and clear country code, cause we cant analyse this type of sales
countries[['region', 'sub-region']] = countries[['region', 'sub-region']].fillna('Antarctica')
countries['alpha-2'] = countries['alpha-2'].fillna('NA')
countries['alpha-3'] = countries['alpha-3'].fillna('NAM')
print(countries[0:154])
events['Units Sold'] = events['Units Sold'].fillna(method = 'ffill')
clean_ev = events.dropna()
print(clean_ev)

numeric_cols = ["Units Sold",'Unit Price',  'Unit Cost' ]
for col in numeric_cols:
    events[col] = pd.to_numeric(events[col], errors="coerce")
events['Order Date'] = pd.to_datetime(events['Order Date'])
events['Ship Date'] = pd.to_datetime(events['Ship Date'])
int_col = events.select_dtypes(include='float64').columns
events[int_col] = events[int_col].astype(int)
text_col = events.select_dtypes(include='object').columns
events[text_col] = events[text_col].astype(str)


def duplicates(df):
  text_columns = df.select_dtypes(include='object').columns
  for col in text_columns:
        events[col] = events[col].astype(str).str.strip().str.lower()
  df = df.drop_duplicates(subset=text_columns)

for col in events.select_dtypes(include='object').columns:
    print(col, events[col].nunique())
events.describe()
events.info()
```

```python
anomalies = {}

#Units Sold
if "Units Sold" in events.columns:
    neg_units = events[events["Units Sold"] <= 0]
    anomalies["negative_or_zero_units"] = neg_units.shape[0]
#Price < Cost
if all(c in events.columns for c in ["Unit Price", "Unit Cost"]):
    bad_margin = events[events["Unit Price"] < events["Unit Cost"]]
    anomalies["price_lower_than_cost"] = bad_margin.shape[0]
#Ship Date before Order Date
if all(c in events.columns for c in ["Ship Date", "Order Date"]):
    bad_dates = events[events["Ship Date"] < events["Order Date"]]
    anomalies["ship_before_order"] = bad_dates.shape[0]
#Outliers (IQR)
if "Units Sold" in events.columns:
    q1 = events["Units Sold"].quantile(0.25)
    q3 = events["Units Sold"].quantile(0.75)
    iqr = q3 - q1
    upper = q3 + 1.5 * iqr
    lower = q1 - 1.5 * iqr
    outliers = events[(events["Units Sold"] > upper) | (events["Units Sold"] < lower)]
    anomalies["units_sold_outliers"] = outliers.shape[0]
print("Anomalies:")
for k, v in anomalies.items():
    print(f"{k} : {v}")
```

```python
# === JOINING THREE TABLES ===
merg_1 = events.merge(products, left_on='Product ID', right_on='id', how='left')
merg_2 = merg_1.merge(countries, left_on='Country Code', right_on='alpha-3', how='left')
merg_2['Sales Channel'] = merg_2['Sales Channel'].str.capitalize()


# === CLEANING AND RENAMING ===
merg_2.drop(['id', 'alpha-2', 'alpha-3'], axis=1, inplace=True)
merg_2.rename(columns={'item_type': 'category', 'name': 'country'}, inplace=True)

# === ADDITIONAL FIELDS ===
merg_2['revenue'] = merg_2['Units Sold'] * merg_2['Unit Price']
merg_2['cost'] = merg_2['Units Sold'] * merg_2['Unit Cost']
merg_2['profit'] = merg_2['revenue'] - merg_2['cost']

# === KEY METRICS ===
total_orders = merg_2['Order ID'].nunique()
total_revenue = merg_2['revenue'].sum()
total_profit = merg_2['profit'].sum()
unique_countries = merg_2['country'].nunique()

print(f"Total Orders: {total_orders}")
print(f"Total Revenue: {total_revenue:,.2f}")
print(f"Total Profit: {total_profit:,.2f}")
print(f"Number of Countries: {unique_countries}")

# === SALES ANALYSIS ===

# By categories
cat_stats = merg_2.groupby('category')[['revenue', 'profit']].sum().sort_values('revenue', ascending=False)
print("\nSales by Categories:")
print(cat_stats.head())

# By countries
country_stats = merg_2.groupby('country')[['revenue', 'profit']].sum().sort_values('revenue', ascending=False)
print("\nSales by Countries:")
print(country_stats.head())

# By channels
channel_stats = merg_2.groupby('Sales Channel')[['revenue', 'profit']].sum()
print("\nSales by Channels:")
print(channel_stats)

# === VISUALIZATIONS ===
import matplotlib.pyplot as plt

# 1. Categories
cat_stats.plot(kind='bar', y='revenue', figsize=(8,5), title='Revenue by Categories')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()

# 2. Countries
country_stats.head(10).plot(kind='barh', y='revenue', figsize=(8,5), title='TOP-10 Countries by Revenue')
plt.xlabel('Revenue')
plt.tight_layout()
plt.show()

# 3. Sales Channels
channel_stats.plot(kind='bar', figsize=(6,4), title='Revenue and Profit by Sales Channels')
plt.ylabel('Amount')
plt.tight_layout()
plt.show()
```

```python
# Delivery time analysis

# Calculate the interval between order and shipment
merg_2["delivery_days"] = (merg_2["Ship Date"] - merg_2["Order Date"]).dt.days

print("\nAverage Delivery Time:", merg_2["delivery_days"].mean().round(2), "days")

# Visualize average delivery time
import matplotlib.pyplot as plt

avg_delivery_by_category = merg_2.groupby("category")["delivery_days"].mean().sort_values()
avg_delivery_by_country = merg_2.groupby("country")["delivery_days"].mean().sort_values()
avg_delivery_by_region = merg_2.groupby("region")["delivery_days"].mean().sort_values()

# 1. By categories
avg_delivery_by_category.plot(kind="bar", figsize=(8,4), title="Average Delivery Time by Category")
plt.ylabel("Days")
plt.tight_layout()
plt.show()

# 2. By countries (TOP 10)
avg_delivery_by_country.head(10).plot(kind="bar", figsize=(8,4), title="Average Delivery Time - TOP 10 Countries")
plt.ylabel("Days")
plt.tight_layout()
plt.show()

# 3. By regions
avg_delivery_by_region.plot(kind="bar", figsize=(8,4), title="Average Delivery Time by Region")
plt.ylabel("Days")
plt.tight_layout()
plt.show()


# PROFIT VS DELIVERY TIME DEPENDENCE

profit_vs_delivery = merg_2.groupby("delivery_days")["profit"].mean().reset_index()

plt.figure(figsize=(7,4))
plt.scatter(profit_vs_delivery["delivery_days"], profit_vs_delivery["profit"], alpha=0.6)
plt.title("Profit vs. Delivery Time Dependence")
plt.xlabel("Number of Delivery Days")
plt.ylabel("Average Profit")
plt.grid(True)
plt.tight_layout()
plt.show()

# SALES DYNAMICS OVER TIME

# Make sure the date is in datetime format
merg_2["Order Date"] = pd.to_datetime(merg_2["Order Date"])

# Add month and year
merg_2["month"] = merg_2["Order Date"].dt.to_period("M")

# Total revenue by month
monthly_sales = merg_2.groupby("month")["revenue"].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind="line", marker="o", title="Sales Dynamics Over Time (Monthly)")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()

# Dynamics by categories
monthly_by_cat = merg_2.groupby(["month", "category"])["revenue"].sum().unstack()

monthly_by_cat.plot(kind="line", figsize=(10,6), title="Sales Dynamics by Categories")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()

# Dynamics by regions
monthly_by_region = merg_2.groupby(["month", "region"])["revenue"].sum().unstack()
monthly_by_region.plot(kind="line", figsize=(10,6), title="Sales Dynamics by Regions")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()


# SALES ANALYSIS BY DAY OF THE WEEK

merg_2["day_of_week"] = merg_2["Order Date"].dt.day_name()

sales_by_day = merg_2.groupby("day_of_week")["revenue"].sum().reindex(
    ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
)

plt.figure(figsize=(8,4))
sales_by_day.plot(kind="bar", color="skyblue", title="Sales by Day of the Week")
plt.ylabel("Revenue")
plt.tight_layout()
plt.show()

# Check for product seasonality (by day of the week)
category_day = merg_2.groupby(["day_of_week", "category"])["revenue"].mean().unstack()

category_day.plot(kind="bar", figsize=(10,6), title="Average Revenue by Day of the Week (by Category)")
plt.ylabel("Average Revenue")
plt.tight_layout()
plt.show()
```
